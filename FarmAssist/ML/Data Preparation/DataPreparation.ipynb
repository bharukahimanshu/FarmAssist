```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
# Reading the data 
crop_data_path = '../Dataset Raw/cropdata.csv'
fertilizer_data_path = '../Dataset Raw/Fertilizer.csv'
crop = pd.read_csv(crop_data_path)
fert = pd.read_csv(fertilizer_data_path)
```

```python
crop
```

```python
fert.head()
```

```python
def change_case(i):
    i = i.replace(" ", "")
    i = i.lower()
    return i
```

```python
fert['Crop'] = fert['Crop'].apply(change_case)
crop['label'] = crop['label'].apply(change_case)
```

```python
fert['Crop'].unique()
```

```python
fert['Crop'] = fert['Crop'].replace('ragi(naachnnii)', 'ragi')
fert['Crop'] = fert['Crop'].replace('barley(jav)', 'barley')
fert['Crop'] = fert['Crop'].replace('rapeseed(mohri)', 'rapeseed')
fert['Crop'] = fert['Crop'].replace('frenchbeans(farasbi)', 'frenchbeans')
fert['Crop'] = fert['Crop'].replace('horsegram(kulthi)', 'horsegram')
fert['Crop'] = fert['Crop'].replace('blackeyedbeans(chawli)', 'blackeyedbeans')
fert['Crop'] = fert['Crop'].replace('limabeans(pavta)', 'limabeans')
fert['Crop'] = fert['Crop'].replace('clusterbeans(gavar)', 'clusterbeans')
fert['Crop'] = fert['Crop'].replace('garciniaindica(kokam)', 'kokam')
fert['Crop'] = fert['Crop'].replace('fenugreekleaf(methi)', 'fenugreekleaf')
fert['Crop'] = fert['Crop'].replace('gooseberry(amla)', 'amla')
fert['Crop'] = fert['Crop'].replace('garciniaindica(kokam)', 'kokam')
fert['Crop'] = fert['Crop'].replace('ziziphusmauritiana(bor)', 'bor')
fert['Crop'] = fert['Crop'].replace('gooseberry(amla)', 'amla')
fert['Crop'] = fert['Crop'].replace('jambun(syzygiumcumini)', 'jambun')
fert['Crop'] = fert['Crop'].replace('jaiphal(nutmeg)', 'jaiphal')
fert['Crop'] = fert['Crop'].replace('greengram(mung)', 'greengram')
fert['Crop'] = fert['Crop'].replace('tapioca(suran)', 'tapoica')
fert['Crop'] = fert['Crop'].replace('blackgram(urad)', 'blackgram')
fert['Crop'] = fert['Crop'].replace('favabeans(papdi-val)', 'favabeans')
fert['Crop'] = fert['Crop'].replace('jowar(sorghum)', 'jowar')
fert['Crop'] = fert['Crop'].replace('mungbeans','mungbean')
fert['Crop'] = fert['Crop'].replace('lentils(masoordal)','lentil')
fert['Crop'] = fert['Crop'].replace('pigeonpeas(toordal)','pigeonpeas')
fert['Crop'] = fert['Crop'].replace('mothbean(matki)','mothbeans')
fert['Crop'] = fert['Crop'].replace('chickpeas(channa)','chickpea')
```

```python
crop.tail()
```

```python
crop_names = crop['label'].unique()
crop_names
```

```python
del fert['Unnamed: 0']
```

```python
crop_names_from_fert = fert['Crop'].unique()
crop_names_from_fert
```

```python
for i in crop_names_from_fert:
    print(crop[crop['label'] == i])
```

```python
extract_labels = []
for i in crop_names_from_fert:
    if i in crop_names:
        extract_labels.append(i)
```

```python
# using extract labesl on crop to get all the data related to those labels
new_crop = pd.DataFrame(columns = crop.columns)
new_fert = pd.DataFrame(columns = fert.columns)
```

```python
for label in extract_labels:
    new_crop = pd.concat([new_crop,crop[crop['label'] == label]])
```

```python
#concat fert to new_fert with unique crop names
for label in extract_labels:
    new_fert = pd.concat([new_fert,fert[fert['Crop'] == label]])
```

```python
new_fert.drop_duplicates(subset = 'Crop', keep = 'first', inplace = True)
```

```python
new_fert
```

```python
#check numeric values in newfert
new_fert.dtypes
#change data type of N,P,K to float
new_fert['N'] = new_fert['N'].astype(float)
new_fert['P'] = new_fert['P'].astype(float)
new_fert['K'] = new_fert['K'].astype(float)
```

```python
import random
temp = pd.DataFrame(columns = ['N','P','K'])
for i in range(0,new_crop.shape[0]):
    crop = new_crop['label'].iloc[i]
    N = new_fert[new_fert['Crop'] == crop]["N"].iloc[0] + random.randint(-20,20)
    P = new_fert[new_fert['Crop'] == crop]["P"].iloc[0] + random.randint(-5,20)
    K = new_fert[new_fert['Crop'] == crop]["K"].iloc[0] + random.randint(-5,5)
    d = {"N":N,"P":P,"K":K}
    temp=pd.concat([temp,pd.DataFrame(d,index=[0])])
```

```python
#drop index
temp.reset_index(drop=True, inplace=True)
```
