```python
import json
import pandas as pd
```

```python
#read json and print
with open('../Dataset Raw/Temperature.json') as f:
    data = json.load(f)
```

```python
df = pd.DataFrame.from_dict(
    {(i, j): data[i][j] for i in data.keys() for j in data[i].keys()}, orient='index')
```

```python
df
```

```python
#df to csv
df.to_csv('../Dataset Processed/Temperature_cleaned.csv')
```
