```python
#importing packages
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
crop_data = pd.read_csv("../Dataset Raw/crop_production.csv")
rainfall_data= pd.read_csv("../Dataset Processed/Rainfall_cleaned.csv")
temp_data= pd.read_csv("../Dataset Processed/Temperature_cleaned.csv")
# creating copy of the dataset to avoid any changes in the original dataset and also avoid the SettingWithCopyWarning
crop = crop_data.copy()
crop.head()
```

```python
#rename yaear to crop_year
crop.rename(columns={'Crop_Year':'Year'}, inplace=True)
```

```python
#strip season values in rainfall_crop
crop['Season'] = crop['Season'].str.strip()
```

```python
#lower the district name in crop data
crop['District_Name'] = crop['District_Name'].str.lower()
```

```python
crop['Yield'] = (crop['Production'] / crop['Area'])
crop.head(10)
```

```python
rainfall_data
```

```python
#sort district name in rainfall_data using district name from crop
crop= crop.sort_values(by=['District_Name'])
```

```python
#drop rainfall data where year is not in crop
rainfall_data = rainfall_data[rainfall_data['Year'].isin(crop['Year'].unique())]
```

```python
rainfall_data['District_Name']=rainfall_data['District_Name'].str.lower()
```

```python
#merge crop with rainfall_data using district and year
rainfall_crop = pd.merge(rainfall_data, crop, on=['District_Name', 'Year'], how='inner')
```

```python
rainfall_crop
```

```python
#create a new column in raincrop for rainfall, take values according to season kharif, rabi, summer , annual, winter using for loop
rainfall_crop['Rainfall'] = np.nan
for i in range(len(rainfall_crop)):
    if rainfall_crop['Season'][i] == 'Kharif':
        rainfall_crop['Rainfall'][i] = rainfall_crop['Kharif'][i]
    elif rainfall_crop['Season'][i] == 'Rabi':
        rainfall_crop['Rainfall'][i] = rainfall_crop['Rabi'][i]
    elif rainfall_crop['Season'][i] == 'Summer':
        rainfall_crop['Rainfall'][i] = rainfall_crop['Summer'][i]
    elif rainfall_crop['Season'][i] == 'Winter':
        rainfall_crop['Rainfall'][i] = rainfall_crop['Winter'][i]
    elif rainfall_crop['Season'][i] == 'Autumn':
        rainfall_crop['Rainfall'][i] = rainfall_crop['Autumn'][i]
    else:
        rainfall_crop['Rainfall'][i] = rainfall_crop['Annual'][i]
```

```python
rainfall_crop
```

```python
#drop Kharif, Rabi, Summer, Winter, Autumn, Annual columns
rainfall_crop = rainfall_crop.drop(['Kharif', 'Rabi', 'Summer', 'Winter', 'Autumn', 'Annual'], axis=1)
```

```python
rainfall_crop['District_Name']=rainfall_crop['District_Name'].str.capitalize()
```

```python
#rearrange columns
rainfall_crop = rainfall_crop[['State_Name', 'District_Name', 'Year', 'Season', 'Crop', 'Area', 'Production', 'Rainfall', 'Yield']]
```

```python
#sort by state name
rainfall_crop = rainfall_crop.sort_values(by=['State_Name','District_Name','Year'])
```

```python
#unique districts
rainfall_crop['District_Name'].nunique()
```

```python
temp_data.drop("Unnamed: 0", axis=1, inplace=True)
temp_data
```

```python
#capitalize district name
temp_data['District_Name']=temp_data['District_Name'].str.capitalize()
```

```python
#drop year in temp_data which are not present in rain crop
temp_data = temp_data[temp_data['Year'].isin(rainfall_crop['Year'].unique())]
```

```python
temp_data['Year'].unique()
```

```python
#merge temp_data with rainfall_crop
final_crop = pd.merge(temp_data, rainfall_crop, on=['District_Name', 'Year'], how='inner')
```

```python
final_crop
```

```python
rainfall_crop.shape
```

```python
#create new column Temperature in final_crop
final_crop['Temperature'] = np.nan
for i in range(len(final_crop)):
    if final_crop['Season'][i] == 'Kharif':
        final_crop['Temperature'][i] = final_crop['Kharif'][i]
    elif final_crop['Season'][i] == 'Rabi':
        final_crop['Temperature'][i] = final_crop['Rabi'][i]
    elif final_crop['Season'][i] == 'Summer':
        final_crop['Temperature'][i] = final_crop['Summer'][i]
    elif final_crop['Season'][i] == 'Winter':
        final_crop['Temperature'][i] = final_crop['Winter'][i]
    elif final_crop['Season'][i] == 'Autumn':
        final_crop['Temperature'][i] = final_crop['Autumn'][i]
    else:
        final_crop['Temperature'][i] = final_crop['Whole Year'][i]
```

```python
# drop Kharif, Rabi, Summer, Winter, Autumn, Annual columns
final_crop = final_crop.drop(['Kharif', 'Rabi', 'Summer', 'Winter', 'Autumn', 'Whole Year'], axis=1)
```

```python
#export data to csv
final_crop.to_csv('../ML/Dataset Processed/crop_production_cleaned.csv', index=False)
```
