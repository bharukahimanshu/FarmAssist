```python
import pandas as pd
```

```python
#read rainfall.csv from data raw
rainfall= pd.read_csv('../Dataset Raw/Rainfall.csv')
rainfall
```

```python
#drop  var, WC14, WC16,WC17.W18
rainfall = rainfall.drop(['VAR', 'WC14', 'WC16', 'WC17', 'WC18'], axis=1)
rainfall
```

```python
rainfall['DIST'].nunique()
```

```python
#drop last 2 rows
rainfall.drop(rainfall.tail(2).index,inplace=True)
```

```python
rainfall
```

```python
#create a new column Summer Rainfall
rainfall['Summer'] = rainfall['Mar'] + rainfall['Apr'] + rainfall['May']+rainfall['Jun']
```

```python
#create a new column Winter Rainfall and add dec of current year and jan and feb of next year
rainfall['Winter'] =rainfall['Dec'] +rainfall['Jan'].shift(-1) + rainfall['Feb'].shift(-1)
```

```python
#create a new column Autumn and consider oct and nov
rainfall['Autumn'] = rainfall['Oct'] + rainfall['Nov']
```

```python
#create a new column Rabi and consider oct, nov, dec
rainfall['Rabi'] = rainfall['Oct'] + rainfall['Nov'] + rainfall['Dec']
```

```python
rainfall.head(10)
```

```python
#rename column WC13 to annual
rainfall.rename(columns={'WC13':'Annual'}, inplace=True)
rainfall.rename(columns={'WC15':'Kharif'}, inplace=True)
```

```python
#create a copy of rainfall
rainfall['Winter'].iloc[-1] = rainfall['Dec'].iloc[-1]
```

```python
rainfall
```

```python
rainfall.isnull().sum()
```

```python
#add +3 to year
rainfall['Year'] = rainfall['Year'] + 3
```

```python
#export to csv
rainfall.to_csv('../Dataset Processed/Rainfall_cleaned.csv', index=False)
```
