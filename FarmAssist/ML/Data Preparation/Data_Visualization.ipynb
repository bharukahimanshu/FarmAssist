```python
#importing packages
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
crop = pd.read_csv("../Dataset Processed/crop_production_cleaned.csv")
crop.head()
```

Visualization

```python
sns.pairplot(crop.sample(1000))
```

```python
#correlation between area,Year, production and yeild
crop.corr(numeric_only=True)
```

```python
sns.heatmap(crop.corr(numeric_only=True), annot=True)
plt.title('Correlation Matrix')
```

```python
#barplot between state and production
sns.boxplot(x=crop['State_Name'],y=crop['Production'])
plt.xticks(rotation=90)
```

```python
sns.boxplot(x=crop["Year"],y= crop["Yield"])
plt.xticks(rotation=90)
```

```python
sns.lineplot(x=crop["Season"], y=crop["Yield"])
plt.xticks(rotation=90)
```

```python
sns.lineplot(x=crop["Season"], y=crop["Production"])
plt.xticks(rotation=90)
```

```python
sns.lineplot(x=crop["Temperature"], y=crop["Production"])
plt.xticks(rotation=90)
```

```python
sns.lineplot(x=crop["Rainfall"], y=crop["Production"])
plt.xticks(rotation=90)
```

```python
sns.barplot(x=crop["Crop"], y=crop["Yield"])
plt.xticks(rotation=90)
plt.tick_params(axis='x', which='major', labelsize=5)
```

Conclusions Obtained from the dataset:
1. Kerela is the largest producer of crop in India.
2. Production per unit area (Yield ) of Puducherry is maximum.
3. In the year 2014, crop yield was maximum.
4. Coconut has highest yield.

```python
#major crops in crop column
crop["Crop"].value_counts()
```

1. Rice

```python
rice = crop[crop["Crop"] == "Rice"]
rice
```

```python
sns.barplot(x="Season", y="Yield", data=rice)
plt.xticks(rotation=90)
```

```python
sns.barplot(x="State_Name", y="Yield", data=rice)
plt.xticks(rotation=90)
```

```python
sns.barplot(x="Year", y="Production", data=rice)
plt.xticks(rotation=90)
```

```python
sns.jointplot(x="Area", y="Production", data=rice, kind="reg")
plt.xticks(rotation=90)
```

Observations :

1. Rice yield is maximum in Rabi season.
2. Rice yield is maximum in Chandigarh.
3. Rice yield has been growing a little from the year 2012 to 2017.
